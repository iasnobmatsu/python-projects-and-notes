```python
import requests

r=requests.get("https://www.rottentomatoes.com/top/bestofrt/")
text=r.text
```

```python
from bs4 import BeautifulSoup


path="https://www.rottentomatoes.com"
soup=BeautifulSoup(text)
for item in soup.select(".table .articleLink"):
    title=item.text.strip()
    href=path+item["href"]
    print(title)
    print(href)
```
